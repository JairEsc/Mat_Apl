```python
import numpy as np
```

```python
n=1000
proportion_as_n=[]
heads=0
for i in range(1,n+1):
    if np.random.uniform(0,1)<0.3:
        heads+=1
    proportion_as_n.append(heads/i)
print(proportion_as_n[-1])
```

```python
import matplotlib.pyplot as plt
plt.plot(proportion_as_n)
plt.xlabel('Valor de $n$')
plt.ylabel('Proporción de éxitos')
```
