```python
import numpy as np
```

```python
n=1000
for p in [0.3,0.03]:
    proportion_as_n=[]
    heads=0
    for i in range(1,n+1):
        if np.random.uniform(0,1)<p:
            heads+=1
        proportion_as_n.append(heads/i)
    print("Aproximacion: ",proportion_as_n[-1])
    import matplotlib.pyplot as plt
    plt.plot(proportion_as_n)
    plt.xlabel('Valor de $n$')
    plt.ylabel('Proporción de éxitos')
    plt.title('Lanzamiento de moneda con $p=$'+str(p))
    plt.show()
```

```python
for n in [1,5,25,100]:
    simulacion=[]
    for j in range(100):
        x_n=0
        for i in range(n):
            x=np.random.uniform(0,1)
            x_n+=x
        simulacion.append(x_n/n)
    plt.plot(simulacion,'p',label='Simulation')
    plt.ylim((.45,.55))
    plt.plot([0,100],[0.5,0.5],'green',label='Theoretical mean')
    plt.plot([0,100],[np.mean(simulacion),np.mean(simulacion)],'red',label='Empirical mean')
    plt.legend(framealpha=1, frameon=True);
    print("Theoretical variance: ",1/(12*n))
    print("Empirical variance: ",np.var(simulacion)) 
    plt.show()
```
