```python
#Proceso de Poisson.
from scipy import stats
def Poisson_arrive(t,l):
    T=0
    t_arrib=[0]
    while(T<t):
        t_arrib.append(t_arrib[-1]+stats.distributions.expon.rvs(0,l))
        T=t_arrib[-1]
    if(T<t):
        return(t_arrib)
    else:
        return(t_arrib[:-1])
import matplotlib.pyplot as plt
import numpy as np
Tiempos_arribo=Poisson_arrive(50,2)
for i in range(len(Tiempos_arribo)-1):
    plt.plot([Tiempos_arribo[i],Tiempos_arribo[i+1]],[i,i],"-b")
Tiempos_arribo[i+1]
plt.show()
#Ejercicio 1 inciso b.
c=50
Tiempos_arribo=Poisson_arrive(c,5)
N=len(Tiempos_arribo)-1
for i in range(N):
    t_0=Tiempos_arribo[i]
    t_f=Tiempos_arribo[i+1]
    t=np.linspace(t_0,t_f,100)
    plt.plot(t,i-(t/c)*N,'b')
t=np.linspace(t_f,c,100)
plt.plot(t,i-(t/c)*N,'b')
```
