```python
#Proceso de Poisson.
from scipy import stats
def Poisson_arrive(t,l):
    T=0
    t_arrib=[0]
    while(T<t):
        t_arrib.append(t_arrib[-1]+stats.distributions.expon.rvs(0,l))
        T=t_arrib[-1]
    if(T<t):
        return(t_arrib)
    else:
        return(t_arrib[:-1])
import matplotlib.pyplot as plt
import numpy as np
Tiempos_arribo=Poisson_arrive(50,2)
plt.plot([Tiempos_arribo[0],Tiempos_arribo[1]],[0,0],"-k")
plt.xlabel(str("Tiempo"),fontsize=11)
plt.ylabel("NÃºmero de eventos")
plt.title("Trayectoria N(t)")
for i in range(1,len(Tiempos_arribo)-1):
    plt.plot([Tiempos_arribo[i],Tiempos_arribo[i+1]],[i,i],"-k")
Tiempos_arribo[i+1]

plt.show()
#Ejercicio 1 inciso b.
c=50
N=len(Tiempos_arribo)-1
for i in range(N):
    t_0=Tiempos_arribo[i]
    t_f=Tiempos_arribo[i+1]
    t=np.linspace(t_0,t_f,100)
    plt.plot(t,i-(t/c)*N,'k')
t=np.linspace(t_f,c,100)
plt.plot(t,i-(t/c)*N,'k')
plt.xlabel("Tiempo")
plt.ylabel("Valor")
plt.title(r"Trayectoria $X(t)=N(t)-\frac{t}{c}N(c)$")
plt.show()
```
