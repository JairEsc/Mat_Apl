```python
N=2000#Poblacion
i_0=300#Frecuencia AA inicial.
G=1000#Numero de generaciones
```

```python
from scipy import stats
import matplotlib.pyplot as plt
Trayectoria=[i_0]
for i in range(G):
    if(i_0==0 | i_0==N):
        print("Se exintguieron")
        break
    else:
        i_0=stats.distributions.binom.rvs(N,i_0/N,1)
        Trayectoria.append(i_0)
plt.plot(Trayectoria)
plt.show()
import matplotlib.pyplot as plt
```

```python
#Con mutacion
alpha=0.01
beta=0.02
#i_0=300
Trayectoria=[i_0]
for i in range(G):
    p_i=(1-alpha)*(i_0/N)+beta*(1-i_0/N)
    i_0=stats.distributions.binom.rvs(N,p_i,1)
    Trayectoria.append(i_0)
plt.plot(Trayectoria)
plt.show()
import matplotlib.pyplot as plt
```
