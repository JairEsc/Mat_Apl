```python
import numpy as np
import matplotlib.pyplot as plt
```

We are ploting $f(x,y)$ in [-1.1,1.1]x[-1.1,1.1], and we define it by using np.linspace() function.

```python
def f(x, y):
    return (x**2+y**2-1)**2+(y**2-1)**2

x = np.linspace(-1.1, 1.1, 30)
y = np.linspace(-1.1, 1.1, 30)
```

We use np.meshgrid() in order to create a set of points where evaluate $f$. The outputs are two list.

```python
X, Y = np.meshgrid(x, y)
```

Now we evaluate the funcion and plot it using a 3D projection.

```python
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
