```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Ejercicio 5-----------------
def f(x,y):
    return(8*x+12*y+x**2-2*y**2)
x = np.linspace(-7,-1, 50)
y = np.linspace(0, 6, 50)

X, Y = np.meshgrid(x, y)
Z=f(X,Y)
plt.contour(X,Y,Z)
plt.xlabel('x_1')
plt.ylabel('x_2     ',rotation=0)
plt.title('Curvas de nivel de $f(x,y)=8 x_{1}+12 x_{2}+x_{1}^{2}-2 x_{2}^{2}$')
```

```python
#Ejercicio 6-----------------
def f(x,y):
    return(100*(y-x**2)**2+(1-x**2))
x = np.linspace(0,3, 50)
y = np.linspace(0, 2, 50)

X, Y = np.meshgrid(x, y)
Z=f(X,Y)
plt.contour(X,Y,Z)
plt.xlabel('x_1')
plt.ylabel('x_2     ',rotation=0)
plt.title('Curvas de nivel de $f(x_1,x_2)=100(x_2-x_1^2)^2+(1-x_1)^2$')
```
