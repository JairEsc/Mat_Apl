```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy.lib.function_base import rot90
def level_set(c):
    x=np.linspace(np.sqrt(12+.00001),10,100)
    y_1=np.sqrt(x**2-12)
    plt.plot(x,y_1,'b',label='$f_1(x_1,x_2)=16$')
    plt.plot(x,-y_1,'b')
    plt.plot(-x,-y_1,'b')
    plt.plot(-x,y_1,'b')
    z_1=np.linspace(2,10)
    z_2=16/(2*z_1)
    plt.plot(z_1,z_2,'g',label='$f_2(x_1,x_2)=12$')
    plt.plot(-z_1,-16/(2*z_1),'g')
    x_p=np.linspace(0.5,2,10)
    plt.plot(x_p,16/(2*x_p),'.g')
    plt.plot(-x_p,-16/(2*x_p),'.g')
    plt.plot([-4,4],[-2,2],'.r',label='$[f_1,f_2]=[12,16]$')
    plt.title('Curvas de nivel')
    plt.ylabel('$x_2$',rotation=0)
    plt.xlabel('$x_1$')
    plt.legend(framealpha=1, frameon=True)
    plt.show()
level_set(0)
```
