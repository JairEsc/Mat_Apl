```python
#Graficas
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-np.pi,np.pi,100)
plt.plot(x,np.sin(x))
plt.plot(x,x)
plt.plot(x,x-(1/6)*x**3)
plt.axis()
plt.legend(["sin(x)",r"$S_1=x$",r"$S_3=x-\frac{x^3}{6}$"])
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,2*np.pi*(3/(2*np.pi**3))*x)
plt.plot(x,2*np.pi*(3/(2*np.pi**3))*x+((4/5)*np.pi*(np.pi**2-15)*(175/(8*np.pi**7)))*(x**3-3*x*np.pi**2/5))
plt.axis()
plt.legend(["sin(x)",r"best $S_1$",r"best $S_3$"])
```
