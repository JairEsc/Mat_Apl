```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def P(a,b,n):
    return (b*n**2/(a**2+n**2))
P_v=np.vectorize(P)
n=np.linspace(0.001,100,1000)
for a in [1,5,10]:
    plt.plot(n,P_v(a=a,b=1,n=n),label='a='+str(a))
plt.legend(framealpha=1, frameon=True);
plt.title('Comparación de tasa de depredación $P(n)/b$')
plt.xlabel('n')
plt.ylabel('P(n)/b            ',rotation=0)
plt.show()
```

```python
from scipy.integrate import odeint
import numpy
import matplotlib.pyplot as plt

t = np.linspace(0,30,2000)
q=35
s=0.4
r=3
for r in [1,3]:
    def derivada(N,tau):
        return (s*r*N*(1-N/q)-N**2/(1+N**2))
    for N_0 in [40,25,2,0.05]:
        sol = odeint(derivada, N_0, t)
        plt.plot(t, sol, label='$N_0=$'+str(N_0))
    plt.legend(framealpha=1, frameon=True);
    plt.title('Solucion $N(t^*)$, para r='+str(r))
    plt.xlabel('t^*')
    plt.ylabel('N     ',rotation=0)
    plt.show()
```
